#Uniform distribution

##Question 1

Suppose $X$ and $Y$ are jointly uniform in the region

$D = \{(x, y): y \leq 2x, 0 < x< 1 , 0 < y <2  \}$

Based on the above information answer the questions (i), (ii) and (iii).

(i) What is the value of $f_{XY}(x, y)$ within the region of support?

$D$ denotes the area of the support$(X, Y)$.

Area of $D$ = $\frac{1}{2} \times 1 \times 2=1$

Since $(X, Y) \sim \text{Uniform}(D)$, it implies that 

$f_{XY}(x, y) = 1, \hspace{1cm}  x, y \in D$

(ii) Find P$(X>Y).$

1_1.png

$\textbf{Solution:}$

1.1 (1).png


$\underline{Method 1:}$ Using area method
 
The shaded region $(A)$ in red represent $(X>Y)$ within the region of support.

Area of $A$ = $\frac{1}{2} \times 1 \times 1=\frac{1}{2}$

 Now, 
 \begin{align*}
     P(X>Y) =& \dfrac{\mid A \mid }{\mid D \mid}\\
     =& \dfrac{1/2}{1} = \dfrac{1}{2}
 \end{align*}

 
 $\underline{Method 2:}$ Using integration method
    \begin{align*}
    P(X>Y) =& \int\limits_{0}^{1}\int\limits_{y}^{1} 1~dxdy\\
    =& \int\limits_{0}^{1} (x)\Biggr|_{y}^{1}dy\\
    =& \int\limits_{0}^{1} (1-y)dy\\
    =&  \left(y-\dfrac{y^2}{2}\right)\Biggr|_{0}^{1}\\
    =& \left(1-\dfrac{1}{2}\right)-\left(0-\dfrac{0}{2}\right)\\
    =&  \dfrac{1}{2}
\end{align*}

(ii) Find P$(X+Y\geq 1).$

$\textbf{Solution:}$
1.2.png

$\underline{Method 1:}$ Using area method
 
 The shaded region in red represent $(X+Y\geq 1)$ within the region of support.

 \begin{align*}
     P(X>Y) =& \dfrac{\mid A \mid }{\mid D \mid}\\
     =& \dfrac{1-\left(\dfrac{1}{2}\times \dfrac{1}{3}\times \dfrac{2}{3}\right)-\left(\dfrac{1}{2}\times \dfrac{2}{3}\times \dfrac{2}{3}\right)}{1} = \dfrac{2}{3}
 \end{align*}

$\underline{Method 2:}$ Using integration method

 \begin{align*}
    P(X+Y\geq 1) =& 1-P(X+Y\leq 1)\\
    =& 1-\int\limits_{0}^{2/3}\int\limits_{y/2}^{1-y} 1~dxdy\\
    =& 1-\int\limits_{0}^{2/3} (x)\Biggr|_{y/2}^{1-y}dy\\
    =& 1-\int\limits_{0}^{2/3} \left(1-\dfrac{3y}{2}\right)dy\\
    =& 1-  \left(y-\dfrac{3y^2}{4}\right)\Biggr|_{0}^{2/3}\\
   =& 1-\dfrac{1}{3}= \dfrac{2}{3}
\end{align*}

#Non-uniform distribution

##Question1
The joint pdf of two random variables $X$ and $Y$ is given by,


$
f_{XY}(x, y)=
\begin{cases}
4xy   & 0 \leq x \leq 1, 0 \leq y \leq 1\\
0  & \text{otherwise}
\end{cases}
$

Find $P(X>2Y)$.

```python
#Joint density function of X and Y

def f1(x,y):      
  return 4*x*y  
```

**3D plot of joint density function** $f_{XY}(x,y)$ 

```python
x = np.linspace(0,1,500)
y = np.linspace(0,1,500) 
X, Y = np.meshgrid(x,y)
fig = plt.figure(figsize =(6,6))
ax = plt.axes(projection = '3d')
# ax.set_xticks([-1, -0.5, 0, 0.5, 1])
# ax.set_yticks([-1, -0.5, 0, 0.5, 1])
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$')
ax.set_zlabel('$Z$')
ax.text2D(0.5, 0.5, "$f_{XY}$", color='red', transform=ax.transAxes, fontsize = 30)
ax.plot_surface(X, Y, f1(X,Y), color = "grey")
```

**3D plot of joint density function** $f_{XY}(x,y)$ 

This is an interactive plot to play with!!!

```python
import plotly.graph_objects as go
df1 = pd.DataFrame(list(zip(x, y, f1(x,y))), columns=['X', 'Y', 'Z'])
fig = go.Figure(data = [go.Surface(z=df1.values,colorscale = 'Greens')])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title = 'Non-uniform Function', autosize= False, width = 500, height =500, margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

$\textbf{Solution:}$


1.1.png$
\begin{align*}
    P(X>2Y) & = \int_{y=0}^{\frac{1}{2}}\int_{x=2y}^{1}f_{XY}(x,y)dxdy \\
    & = \int_{y=0}^{\frac{1}{2}}\int_{x=2y}^{1}4xy~dxdy \\
    & = 4\int_{y=0}^{\frac{1}{2}}y\left(\int_{x=2y}^{1}4x~dx\right)dy \\
     & = 8\int_{y=0}^{\frac{1}{2}}y (x^2)\biggr|_{2y}^{1} dy \\
     &= 8\int_{y=0}^{\frac{1}{2}} y(1-4y^2)dy \\
     &= 8\left(\dfrac{y^2}{2}-\dfrac{4y^4}{4}\right)\Biggr|_{0}^{\frac{1}{2}} \\
     &= 8\left(\dfrac{1}{8}-\dfrac{1}{16}\right) = \dfrac{1}{2}
\end{align*}
$

The blue shaded region represents $X-Y<1$ 
inside the region of support.


1_2.1.png$\begin{align*}
    P(X-Y<1)& = 1-P(X-Y\geq 1)\\                    
    &=1-\int_{x=1}^{2}\int_{y=0}^{x-1}f_{XY}(x,y)dydx \\
    & = 1-\dfrac{1}{4}\int_{x=1}^{2}\int_{y=0}^{x-1}(x+y)~dydx \\
    & = 1-\dfrac{1}{4}\int_{x=1}^{2}\left(xy+\dfrac{y^2}{2}\right )\biggr|_{y=0}^{x-1}~dx\\

\end{align*}

##Question 3

The joint pdf of two random variables $X$ and $Y$ is given by

$
f_{XY}(x, y)=
\begin{cases}
\dfrac{3}{4}+xy  & 0 \leq x \leq 1, 0 \leq y \leq 1\\
0  & \text{otherwise}
\end{cases}
$

Find the value of $P\left(Y>\dfrac{1}{2}|X=\dfrac{1}{4}\right)$.

**3D plot of joint density function** $f_{XY}(x,y)$ 

```python
def f3(x,y):
  return 0.75+ x*y
```

```python
x = np.linspace(0,1,500)
y = np.linspace(0,1,500) 
X, Y = np.meshgrid(x,y)
fig = plt.figure(figsize = (6,6))
ax = plt.axes(projection = '3d')

ax.set_xlabel('$X$')
ax.set_ylabel('$Y$')
ax.set_zlabel('$Z$')
ax.text2D(0.5, 0.5, "$f_{XY}$", color='red', transform=ax.transAxes, fontsize = 30)
ax.plot_wireframe(X, Y, f3(X,Y), color = "grey")
```

$\textbf{Solution:}$

We know that, 
$
f_{Y|X=x}(y) = \dfrac{f_{XY}(x, y)}{ f_{X}(x)} \hspace{15mm} \ldots (1)
$

Now,

$ \begin{align*}
f_X(x) =& \int\limits_{0}^{1} f_{XY}(x, y)~dy\\
=& \int\limits_{0}^{1} \left(\dfrac{3}{4}+xy\right) ~dy\\
=& \left( \dfrac{3y}{4}+\dfrac{xy^2}{2} \right) \Biggr|_{0}^{1}\\
=& \left( \dfrac{3}{4}+\dfrac{x}{2} \right) 
\end{align*}$

On substituting the value of $f_X(x)$ in equation (1), we get 

$
f_{Y|X=x}(y) = \dfrac{\left(\dfrac{3}{4}+xy \right)}{\left(\dfrac{3}{4}+\dfrac{x}{2} \right)} 
f_{Y|X=x}(y) = \dfrac{3+4xy}{3+2x} $

$\implies f_{Y|X=\frac{1}{4}}(y) = \dfrac{3+y}{\left(\frac{7}{2}\right) } 
$

$
 \begin{align*}
P\left(Y>\dfrac{1}{2}|X=\dfrac{1}{4}\right) =& \int\limits_{\frac{1}{2}}^{1} \left(\dfrac{3+y}{\left(\frac{7}{2}\right)}\right)~dy\\
=& \dfrac{2}{7} \int\limits_{\frac{1}{2}}^{1} \left(3+y)\right) ~dy\\
=& \dfrac{2}{7}  \left( 3y+\dfrac{y^2}{2} \right) \Biggr|_{\frac{1}{2}}^{1}\\
=& \dfrac{2}{7} \left[\left(3+\dfrac{1}{2} \right) - \left(\dfrac{3}{2}+\dfrac{1}{8} \right) \right] \\
=& \dfrac{2}{7} \left[\left(\dfrac{7}{2} \right) - \left(\dfrac{13}{8} \right) \right] \\
=& \dfrac{2}{7} \left( \dfrac{15}{8}\right)\\
=& \dfrac{15}{28}
\end{align*}$

##Question 4

 Let $X$ and $Y$ be jointly distributed with pdf
$
f_{XY}(x, y)=
\begin{cases}
\dfrac{1}{4}(1+xy)  &  |x| < 1, |y| < 1\\
0  & \text{otherwise}
\end{cases}
$

Are $X^2$ and $Y^2$ independent?

**3D plot of joint density function** $f_{XY}(x,y)$ 

```python
def f4(x,y):
  return 0.25*(1+ x*y)
```

```python
x = np.linspace(-1,1,500)
y = np.linspace(-1,1,500) 
X, Y = np.meshgrid(x,y)
fig = plt.figure(figsize = (6,6))
ax = plt.axes(projection = '3d')
ax.set_xticks([-1, -0.5, 0, 0.5, 1])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$')
ax.set_zlabel('$Z$')
ax.text2D(0.5, 0.5, "$f_{XY}$", color='red', transform=ax.transAxes, fontsize = 30)
ax.plot_wireframe(X, Y, f4(X,Y), color = "grey")
```

**3D plot of joint density function** $f_{XY}(x,y)$ 

This is an interactive plot to play with!!!

```python
import plotly.graph_objects as go
df4 = pd.DataFrame(list(zip(x, y, f4(x,y))), columns=['X', 'Y', 'Z'])
fig = go.Figure(data = [go.Surface(z=df4.values,colorscale = 'Blues')])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title = 'Non-uniform Function', autosize= False, width = 500, height =500, margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

$\textbf{Solution:}$
 \begin{align*}
f_X(x) =& \int\limits_{-1}^{1} f_{XY}(x, y)~dy\\
=& \int\limits_{-1}^{1} \dfrac{1}{4}(1+xy)~dy\\
=& \dfrac{1}{4}\left(y+\dfrac{xy^2}{2} \right) \Biggr|_{-1}^{1}\\
=& \dfrac{1}{4} \left[\left(1+\dfrac{x}{2} \right) - \left (-1+\dfrac{x}{2} \right)\right]\\
=& \dfrac{1}{2}, \hspace{4mm}-1 < x < 1
\end{align*}
Similarly, \begin{align*}
f_Y(y) =& \int\limits_{-1}^{1} f_{XY}(x, y)~dx\\
=& \int\limits_{-1}^{1} \dfrac{1}{4}(1+xy)~dx\\
=& \dfrac{1}{4}\left(x+\dfrac{yx^2}{2} \right) \Biggr|_{-1}^{1}\\
=& \dfrac{1}{4} \left[\left(1+\dfrac{y}{2} \right) - \left (-1+\dfrac{y}{2} \right)\right]\\
=& \dfrac{1}{2}, \hspace{4mm}-1 < y < 1\\
\end{align*}
Now, 
\begin{align*}
P(X^2 \leq x) =& P(|X| \leq \sqrt{x})\\
              =& \int\limits_{-\sqrt{x}}^{\sqrt{x}} f_{X}(x)~dx\\
              =& \int\limits_{-\sqrt{x}}^{\sqrt{x}} \dfrac{1}{2}~dx\\
              =& \dfrac{1}{2} (x)\Biggr|_{-\sqrt{x}}^{\sqrt{x}}\\
              =& \dfrac{1}{2}\left(\sqrt{x}-(-\sqrt{x}) \right)\\
              =& \sqrt{x}
\end{align*}
Similarly, 
\begin{align*}
P(Y^2 \leq y) =& P(|Y| \leq \sqrt{y})\\
              =& \int\limits_{-\sqrt{y}}^{\sqrt{y}} f_{Y}(y)~dy\\
               =& \sqrt{y}
\end{align*}

Now, If $X^2$ and $Y^2$ to be independent then the products of their marginal densities should be equal to their joint density.

\begin{align*}
P(X^2 \leq x \cap Y^2 \leq y) =& P(|X| \leq \sqrt{x} \cap |Y| \leq \sqrt{y})\\
              =& \int\limits_{-\sqrt{y}}^{\sqrt{y}} \int\limits_{-\sqrt{x}}^{\sqrt{x}} f_{XY}(x,y)~dxdy\\
              =& \int\limits_{-\sqrt{y}}^{\sqrt{y}}\left( \int\limits_{-\sqrt{x}}^{\sqrt{x}} \dfrac{1}{4}(1+xy)~dx\right)dy\\
              =& \dfrac{1}{4} \int\limits_{-\sqrt{y}}^{\sqrt{y}} \left( x + \dfrac{yx^2}{2}\right)\Biggr|_{-\sqrt{x}}^{\sqrt{x}} ~dy\\
               =& \dfrac{1}{4} \int\limits_{-\sqrt{y}}^{\sqrt{y}} 2 \sqrt{x}~dy\\
               =&  \dfrac{\sqrt{x}}{2} \int\limits_{-\sqrt{y}}^{\sqrt{y}}  1~dy\\
               =& \dfrac{\sqrt{x}}{2} (y)\Biggr|_{-\sqrt{y}}^{\sqrt{y}}\\
               =& \dfrac{\sqrt{x}}{2} \times( 2\sqrt{y})\\
               =& \sqrt{x} \sqrt{y}\\
               =& P(X^2 \leq x)  P(Y^2 \leq y) 
\end{align*}

Thus $X^2$ and $Y^2$ are independent.

##Question 5

The joint density function of the random variables $X$ and $Y$ is given by 
$
f_{XY}(x, y)=
\begin{cases}
\dfrac{x+y}{k}  & 0 < x < 2, 0 < y < 2\\
0  & \text{otherwise}
\end{cases}
$

Calculate the value of $k$.

**3D plot of joint density function** $f_{XY}(x,y)$ 

```python
#Joint density of X and Y
def f5(x,y):
  return (x+y)/8
```

```python
x = np.linspace(0,2,500)
y = np.linspace(0,2,500) 
X, Y = np.meshgrid(x,y)
fig = plt.figure(figsize = (6,6))
ax = plt.axes(projection = '3d')
ax.set_xticks([0, 0.5, 1, 1.5, 2])
ax.set_yticks([0, 0.5, 1, 1.5, 2])
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$')
ax.set_zlabel('$Z$')
ax.text2D(0.5, 0.5, "$f_{XY}$", color='red', transform=ax.transAxes, fontsize = 25)
ax.plot_wireframe(X, Y, f5(X,Y), color = "grey")
```

**3D plot of joint density function** $f_{XY}(x,y)$

```python
import plotly.graph_objects as go
df5 = pd.DataFrame(list(zip(x, y, f5(x,y))), columns=['X', 'Y', 'Z'])
fig = go.Figure(data = [go.Surface(z=df5.values,colorscale = 'Blues')])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title = 'Non-uniform Function', autosize= False, width = 500, height =500, margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

$\textbf{Solution:}$

 question_5.png$\int\limits_{y=-\infty}^{y=\infty} \int\limits_{x=-\infty}^{x=\infty} f_{XY}(x,y)~dxdy = 1$

   $\implies \int\limits_{y=0}^{y=2} \int\limits_{x=0}^{x=2} \dfrac{x+y}{k}~dxdy = 1$

  $\implies \int\limits_{y=0}^{y=2} \left( \dfrac{x^2}{2}+xy \right)\Biggr|_{0}^{2} ~dy = k$

    
$ \implies k=  \int\limits_{y=0}^{y=2} (2+2y)~dy$

$\implies k= \left(2y+\dfrac{2y^2}{2} \right)\Biggr|_{0}^{2} $

$\implies k= 2\times 2+\dfrac{2\times(2)^2}{2} $

$ \implies k=8$


By using the pdf in above example, calculate the value of $P(0<X<1, 0<Y<2)$?

$\begin{align*}
      P(0<X<1, 0<Y<2) =& \int\limits_{y=0}^{y=2} \int\limits_{x=0}^{x=1} \dfrac{x+y}{8}~dxdy\\
      =& \dfrac{1}{8} \int\limits_{y=0}^{y=2} \left( \dfrac{x^2}{2}+xy \right)\Biggr|_{0}^{1} ~dy \\
      =& \dfrac{1}{8} \int\limits_{y=0}^{y=2} \left( \dfrac{1}{2}+y \right) ~dy \\
       =& \dfrac{1}{8} \left( \dfrac{y}{2}+\dfrac{y^2}{2} \right)\Biggr|_{0}^{2}\\
        =& \dfrac{1}{8} \left( \dfrac{2}{2}+\dfrac{2^2}{2} \right)\\
        =& \dfrac{3}{8}
\end{align*}$

**From the 2D plot, it looks like the probabilty is $\dfrac{1}{2}$, but after the integration the answer is different. It is happening since the joint pdf is non uniform. To understand or visualise better one can refer to 3D plot.**
